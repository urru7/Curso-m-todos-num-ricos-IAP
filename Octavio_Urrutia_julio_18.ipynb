**Intepretación de la tabla**:
En esta sección se tomaron el conjunto de datos de las viviendas en y se dividieron en subconjuntos de entrenamiento y prueba. Los datos están también mezclados en un orden aleatorio cuando, para eliminar cualquier parcialidad en el ordenamiento del conjunto de datos.

```python
from sklearn.impute import SimpleImputer
```

Esta linea es algo similar a una función que completa o imputa los datos

```python
llenador_de_datos_faltantes = SimpleImputer(strategy = 'most_frequent') 
```

En esta linea se constituyo un llenador de datos faltates utilizando la moda 

```python
v_imputados = llenador_de_datos_faltantes.fit_transform(v)
```

 Esta linea de código se utlizó para entrenar y transformar todos los datos.

```python
v_imputados_df = pd.DataFrame(v_imputados, columns = v.columns, index = v.index)  
```

Con esta linea se pretende convertir la información a un dataframe para poder visualizar si los datos se rellenaron correctamente. 

```python
v_imputados_df.info() 
```

En la tabla se observa que los datos de los dormitorios ya estan completos.
